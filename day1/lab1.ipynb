```python

messages=[
  {"role": "system", "content": "You are a poetic assistant, skilled in explaining complex programming concepts with creative flair."},
  {"role": "user", "content": "Compose a poem that explains the concept of recursion in programming."}
]

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=messages
)

print(completion.choices[0].message)
```

```python
from llmlingua import PromptCompressor

structured_prompt = """I have both a personal and work ssh on my mac, and in a situation where I use the work ssh as the main one, please tell me the command to use the personal ssh intermittently to commit."""


llm_lingua = PromptCompressor()
compressed_prompt = llm_lingua.compress_prompt(structured_prompt, instruction="", question="", target_token=200)

print(compressed_prompt)
```
